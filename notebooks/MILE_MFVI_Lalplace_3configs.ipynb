#MFVI

```python
import jax
import jax.numpy as jnp
import optax
from flax.training.train_state import TrainState
from jax.flatten_util import ravel_pytree

X_train = loader.train_x
y_train = loader.train_y
X_val = loader.valid_x
y_val = loader.valid_y
X_test = features
y_test = labels

print("Train:", X_train.shape, y_train.shape)
print("Valid:", X_val.shape, y_val.shape)
print("Test:", X_test.shape, y_test.shape)
```

```python
params_init = module.init(config.jax_rng, X_train[:1])

def ce_loss(params, x, y):
    logits = module.apply(params, x, train=True)
    log_probs = jax.nn.log_softmax(logits, axis=-1)
    return -jnp.mean(log_probs[jnp.arange(y.shape[0]), y])

@jax.jit
def train_step(state, x, y):
    loss, grads = jax.value_and_grad(ce_loss)(state.params, x, y)
    return state.apply_gradients(grads=grads), loss

@jax.jit
def eval_point(params, x, y):
    logits = module.apply(params, x, train=False)
    probs = jax.nn.softmax(logits, axis=-1)
    acc = jnp.mean(jnp.argmax(probs, axis=-1) == y)
    lppd = jnp.mean(jnp.log(probs[jnp.arange(y.shape[0]), y] + 1e-12))
    return acc, lppd

state = TrainState.create(
    apply_fn=module.apply,
    params=params_init,
    tx=optax.adam(1e-3),
)

rng = jax.random.key(123)
batch_size = 64
epochs = 300

for epoch in range(epochs):
    perm = jax.random.permutation(rng, X_train.shape[0])
    rng, _ = jax.random.split(rng)
    Xs, ys = X_train[perm], y_train[perm]

    for start in range(0, X_train.shape[0], batch_size):
        state, loss = train_step(
            state,
            Xs[start:start+batch_size],
            ys[start:start+batch_size],
        )

    if epoch % 50 == 0 or epoch == epochs - 1:
        val_acc, val_lppd = eval_point(state.params, X_val, y_val)
        print(f"epoch {epoch:3d} | loss {float(loss):.4f} | val acc {float(val_acc):.4f} | val lppd {float(val_lppd):.4f}")

map_params = state.params
test_acc, test_lppd = eval_point(map_params, X_test, y_test)
print("\nMAP Test | LPPD:", float(test_lppd), "| ACC:", float(test_acc))
```

```python
def metrics_from_sample_probs(sample_probs, y):
    mean_probs = jnp.mean(sample_probs, axis=0)
    pred = jnp.argmax(mean_probs, axis=-1)

    acc = jnp.mean(pred == y)
    true_probs = mean_probs[jnp.arange(y.shape[0]), y]
    lppd = jnp.mean(jnp.log(true_probs + 1e-12))
    nll = -lppd

    entropy = -jnp.mean(jnp.sum(mean_probs * jnp.log(mean_probs + 1e-12), axis=-1))
    sample_entropy = -jnp.mean(
        jnp.sum(sample_probs * jnp.log(sample_probs + 1e-12), axis=-1)
    )
    mi = entropy - sample_entropy

    return {
        "lppd": lppd,
        "nll": nll,
        "acc": acc,
        "predictive_entropy": entropy,
        "mutual_information": mi,
    }

def print_metrics(name, metrics):
    print(f"\n{name} Performance")
    for k, v in metrics.items():
        print(f"{k:22s}: {float(v):.6f}")
```

```python
def split_last_layer(params):
    p = params["params"]
    last_key = list(p.keys())[-1]
    base = {k: v for k, v in p.items() if k != last_key}
    last = {last_key: p[last_key]}
    return {"params": base}, {"params": last}, last_key

base_params, last_params, last_key = split_last_layer(map_params)
flat_last, unravel_last = ravel_pytree(last_params)

D = flat_last.shape[0]
print("Last layer:", last_key)
print("MFVI parameters:", D)

def combine_params(base_params, last_params):
    return {"params": {**base_params["params"], **last_params["params"]}}

def forward_from_vec(theta_vec, x):
    last_p = unravel_last(theta_vec)
    full_p = combine_params(base_params, last_p)
    return module.apply(full_p, x, train=False)

def log_prior(theta_vec, prior_scale=1.0):
    return -0.5 * jnp.sum((theta_vec / prior_scale) ** 2)

def log_likelihood(theta_vec, x, y):
    logits = forward_from_vec(theta_vec, x)
    log_probs = jax.nn.log_softmax(logits, axis=-1)
    return jnp.sum(log_probs[jnp.arange(y.shape[0]), y])

def elbo(var_params, x, y, rng, num_samples=8):
    mean = var_params["mean"]
    log_std = var_params["log_std"]
    std = jnp.exp(log_std)

    def one_sample(key):
        eps = jax.random.normal(key, mean.shape)
        theta = mean + std * eps

        ll = log_likelihood(theta, x, y)
        lp = log_prior(theta)

        dim = mean.shape[0]
        lq = (
            -0.5 * jnp.sum(((theta - mean) / std) ** 2)
            - jnp.sum(log_std)
            - 0.5 * dim * jnp.log(2.0 * jnp.pi)
        )
        return ll + lp - lq

    keys = jax.random.split(rng, num_samples)
    return jnp.mean(jax.vmap(one_sample)(keys))

optimizer = optax.adam(1e-3)
var_params = {
    "mean": flat_last,
    "log_std": jnp.full_like(flat_last, -5.0),
}
opt_state = optimizer.init(var_params)

@jax.jit
def mfvi_step(var_params, opt_state, x, y, rng):
    loss_fn = lambda vp: -elbo(vp, x, y, rng, num_samples=8)
    loss, grads = jax.value_and_grad(loss_fn)(var_params)
    updates, opt_state = optimizer.update(grads, opt_state)
    var_params = optax.apply_updates(var_params, updates)
    return var_params, opt_state, loss

rng = jax.random.key(456)
epochs = 500
batch_size = 64

for epoch in range(epochs):
    perm = jax.random.permutation(rng, X_train.shape[0])
    rng, _ = jax.random.split(rng)
    Xs, ys = X_train[perm], y_train[perm]

    for start in range(0, X_train.shape[0], batch_size):
        rng, subkey = jax.random.split(rng)
        var_params, opt_state, loss = mfvi_step(
            var_params,
            opt_state,
            Xs[start:start+batch_size],
            ys[start:start+batch_size],
            subkey,
        )

    if epoch % 100 == 0 or epoch == epochs - 1:
        print(f"MFVI epoch {epoch:4d} | neg ELBO {float(loss):.4f}")
```

```python
@jax.jit
def probs_for_theta(theta_vec, x):
    logits = forward_from_vec(theta_vec, x)
    return jax.nn.softmax(logits, axis=-1)

S = 100
rng = jax.random.key(789)

eps = jax.random.normal(rng, (S, D))
theta_samples = var_params["mean"][None, :] + jnp.exp(var_params["log_std"])[None, :] * eps

sample_probs_mfvi = jax.vmap(lambda th: probs_for_theta(th, X_test))(theta_samples)

mfvi_metrics = metrics_from_sample_probs(sample_probs_mfvi, y_test)
print_metrics("Last-Layer MFVI", mfvi_metrics)
```

#Laplace

```python
import jax
import jax.numpy as jnp
from jax.flatten_util import ravel_pytree

# Wir nutzen dieselbe Zerlegung wie bei MFVI:
# base_params, last_params, last_key, flat_last, unravel_last, D
print("Last layer:", last_key)
print("Laplace parameters:", D)

def nll_single(theta_vec, x_single, y_single):
    logits = forward_from_vec(theta_vec, x_single[None, :])
    log_probs = jax.nn.log_softmax(logits, axis=-1)
    return -log_probs[0, y_single]

def grad_logits(theta_vec, x_single):
    def logits_fn(th):
        return forward_from_vec(th, x_single[None, :])[0]
    return jax.jacrev(logits_fn)(theta_vec)   # shape: (C, D)

def softmax_hessian_logits(theta_vec, x_single):
    logits = forward_from_vec(theta_vec, x_single[None, :])[0]
    p = jax.nn.softmax(logits)
    return jnp.diag(p) - jnp.outer(p, p)      # shape: (C, C)

@jax.jit
def ggn_one(theta_vec, x_single, y_single):
    J = grad_logits(theta_vec, x_single)      # C x D
    H = softmax_hessian_logits(theta_vec, x_single)  # C x C
    return J.T @ H @ J                        # D x D

def compute_laplace_precision(theta_map, X, y, prior_precision=1.0, max_points=None):
    if max_points is not None:
        X = X[:max_points]
        y = y[:max_points]

    precision = prior_precision * jnp.eye(theta_map.shape[0])

    for i in range(X.shape[0]):
        precision = precision + ggn_one(theta_map, X[i], y[i])
        if i % 100 == 0:
            print(f"GGN point {i}/{X.shape[0]}")

    return precision

# Für ersten Test ggf. nur 700 Trainingspunkte, weil Train hier klein ist
theta_map = flat_last

laplace_precision = compute_laplace_precision(
    theta_map,
    X_train,
    y_train,
    prior_precision=1.0,
    max_points=None,
)

print("Precision shape:", laplace_precision.shape)
```

```python
def sample_laplace(theta_map, precision, rng, n_samples=100, jitter=1e-4):
    D = theta_map.shape[0]
    precision_jitter = precision + jitter * jnp.eye(D)

    L = jnp.linalg.cholesky(precision_jitter)

    eps = jax.random.normal(rng, (n_samples, D))

    # sample = mean + precision^{-1/2} eps
    # solve L.T z = eps.T
    z = jax.scipy.linalg.solve_triangular(
        L.T,
        eps.T,
        lower=False,
    ).T

    return theta_map[None, :] + z

S = 100
rng = jax.random.key(987)

theta_samples_laplace = sample_laplace(
    theta_map,
    laplace_precision,
    rng,
    n_samples=S,
    jitter=1e-3,
)

print(theta_samples_laplace.shape)
```

```python
sample_probs_laplace = jax.vmap(
    lambda th: probs_for_theta(th, X_test)
)(theta_samples_laplace)

laplace_metrics = metrics_from_sample_probs(sample_probs_laplace, y_test)

print_metrics("Last-Layer Laplace", laplace_metrics)
```

```python
#Extension Mile
%cd /content/MILE

import json
import pickle
import re
from pathlib import Path

import jax
import jax.numpy as jnp
import numpy as np

from src.config.core import Config
from src.dataset import TextLoader
from src.dataset.utils import CustomBPETokenizer
import src.training.utils as train_utils
from src.inference.evaluation import evaluate_bde


# ==================================================
# ANPASSEN
# ==================================================

EXP_DIR = Path("/content/drive/MyDrive/MILE_RESULTS/ag_news_test/ag_news_mclmc_small_gpu")

MAX_TEST_POINTS = 1000
MAX_SAMPLES_PER_CHAIN = 512
BATCH_SIZE = 64
N_BINS_ECE = 15

# ==================================================
# Hilfsfunktionen
# ==================================================

def to_float(x):
    try:
        return float(x)
    except Exception:
        return float(np.asarray(x))


def parse_runtime_and_sampler(log_path):
    out = {
        "total_time": None,
        "warmstart_time": None,
        "bde_time": None,
    }

    if not log_path.exists():
        return out

    text = log_path.read_text()

    m = re.search(r"time\.warmstart took ([0-9.]+) seconds", text)
    if m:
        out["warmstart_time"] = float(m.group(1))

    m = re.search(r"time\.sampling took ([0-9.]+) seconds", text)
    if m:
        out["bde_time"] = float(m.group(1))

    if out["warmstart_time"] is not None and out["bde_time"] is not None:
        out["total_time"] = out["warmstart_time"] + out["bde_time"]

    return out


def parse_mclmc_params(warmup_path):
    out = {"step_size": None, "L": None}

    if not warmup_path.exists():
        return out

    lines = warmup_path.read_text().strip().splitlines()
    if len(lines) >= 2:
        step_sizes = np.array([float(x) for x in lines[0].split(",")])
        Ls = np.array([float(x) for x in lines[1].split(",")])
        out["step_size"] = float(step_sizes.mean())
        out["L"] = float(Ls.mean())

    return out


def compute_extra_classification_metrics(logits, labels, n_bins=15):
    """
    Erwartet logits/probs aus evaluate_bde.
    Falls logits bereits Wahrscheinlichkeiten sind, erkennt der Code das grob.
    """

    arr = jnp.asarray(logits)

    # Falls Shape z.B. (N, C), direkt verwenden.
    # Falls Shape mehr Dimensionen hat, versuchen wir über Sample/Chain-Achsen zu mitteln.
    if arr.ndim > 2:
        arr = arr.reshape((-1,) + arr.shape[-2:])  # (S, N, C)
        probs_samples = jax.nn.softmax(arr, axis=-1)
        probs = probs_samples.mean(axis=0)
    else:
        probs_samples = None
        probs = jax.nn.softmax(arr, axis=-1)

    # Wenn bereits Wahrscheinlichkeiten: Summe ungefähr 1 und Werte in [0,1]
    if jnp.all(probs >= 0) and jnp.allclose(probs.sum(axis=-1), 1.0, atol=1e-3):
        pass

    labels = jnp.asarray(labels).reshape(-1)
    n_classes = probs.shape[-1]
    y_onehot = jax.nn.one_hot(labels, n_classes)

    pred = jnp.argmax(probs, axis=-1)
    conf = jnp.max(probs, axis=-1)
    correct = (pred == labels)

    acc = jnp.mean(correct)
    brier = jnp.mean(jnp.sum((probs - y_onehot) ** 2, axis=-1))

    eps = 1e-8
    pred_entropy = -jnp.sum(probs * jnp.log(probs + eps), axis=-1)
    predictive_entropy = jnp.mean(pred_entropy)

    if probs_samples is not None:
        ent_samples = -jnp.sum(probs_samples * jnp.log(probs_samples + eps), axis=-1)
        expected_entropy = jnp.mean(ent_samples)
        mutual_information = predictive_entropy - expected_entropy
    else:
        expected_entropy = np.nan
        mutual_information = np.nan

    # ECE
    ece = 0.0
    bin_edges = jnp.linspace(0.0, 1.0, n_bins + 1)

    for i in range(n_bins):
        lo, hi = bin_edges[i], bin_edges[i + 1]
        mask = (conf > lo) & (conf <= hi)
        prop = jnp.mean(mask)

        bin_acc = jnp.where(jnp.any(mask), jnp.mean(correct[mask]), 0.0)
        bin_conf = jnp.where(jnp.any(mask), jnp.mean(conf[mask]), 0.0)

        ece += prop * jnp.abs(bin_acc - bin_conf)

    return {
        "ece": float(ece),
        "brier": float(brier),
        "predictive_entropy": float(predictive_entropy),
        "expected_entropy": float(expected_entropy),
        "mutual_information": float(mutual_information),
    }


# ==================================================
# Config, Daten, Samples laden
# ==================================================

config = Config.from_file(EXP_DIR / "config.yaml")
tree_dir = EXP_DIR / "tree"

module = config.get_flax_model()

tokenizer = CustomBPETokenizer(
    config.model.context_len,
    max_vocab_size=config.model.vocab_size,
)

loader = TextLoader(
    config=config.data,
    rng=config.jax_rng,
    tokenizer=tokenizer,
    context_len=config.model.context_len,
    omit_freq=100,
    target_len=config.data.target_len,
)

features = loader.test_x[:MAX_TEST_POINTS]
labels = loader.test_y[:MAX_TEST_POINTS]

samples = train_utils.load_samples_from_dir(EXP_DIR / "samples", tree_path=tree_dir)

samples_reduced = jax.tree.map(
    lambda x: x[:, :MAX_SAMPLES_PER_CHAIN, ...],
    samples,
)

print("Experiment:", EXP_DIR.name)
print("Test subset:", features.shape, labels.shape)
print("Sample leaf shape:", jax.tree.leaves(samples_reduced)[0].shape)


# ==================================================
# BDE / MILE Evaluation
# ==================================================

metrics = {}

logits, metrics = evaluate_bde(
    params=samples_reduced,
    module=module,
    features=features,
    labels=labels,
    task=config.data.task,
    batch_size=BATCH_SIZE,
    verbose=False,
    metrics_dict=metrics,
    nominal_coverages=[0.5, 0.75, 0.9, 0.95],
)

extra = compute_extra_classification_metrics(
    logits=logits,
    labels=labels,
    n_bins=N_BINS_ECE,
)

runtime = parse_runtime_and_sampler(EXP_DIR / "training.log")
sampler_params = parse_mclmc_params(EXP_DIR / "warmup_params.txt")


# ==================================================
# Minimal Report bauen
# ==================================================

report = {
    "experiment_name": EXP_DIR.name,
    "max_test_points": MAX_TEST_POINTS,
    "max_samples_per_chain": MAX_SAMPLES_PER_CHAIN,
    "batch_size": BATCH_SIZE,

    "runtime_seconds_total": runtime["total_time"],
    "runtime_seconds_warmstart": runtime["warmstart_time"],
    "runtime_seconds_sampling": runtime["bde_time"],

    "step_size": sampler_params["step_size"],
    "L": sampler_params["L"],

    "accuracy": to_float(metrics["acc"]),
    "lppd": to_float(metrics["lppd"]),
    "nll": to_float(metrics["nll"]),

    "ece": extra["ece"],
    "brier": extra["brier"],
    "predictive_entropy": extra["predictive_entropy"],
    "expected_entropy": extra["expected_entropy"],
    "mutual_information": extra["mutual_information"],
}

print("\n" + "=" * 60)
print("MINIMAL MILE REPORT")
print("=" * 60)

for k, v in report.items():
    if isinstance(v, float):
        print(f"{k:30s}: {v:.6f}")
    else:
        print(f"{k:30s}: {v}")


# ==================================================
# Speichern
# ==================================================

out_json = EXP_DIR / f"minimal_report_test{MAX_TEST_POINTS}_samples{MAX_SAMPLES_PER_CHAIN}.json"
out_pkl = EXP_DIR / f"minimal_report_test{MAX_TEST_POINTS}_samples{MAX_SAMPLES_PER_CHAIN}.pkl"

with open(out_json, "w") as f:
    json.dump(report, f, indent=2)

with open(out_pkl, "wb") as f:
    pickle.dump(report, f)

print("\nSaved JSON:", out_json)
print("Saved PKL :", out_pkl)
```
