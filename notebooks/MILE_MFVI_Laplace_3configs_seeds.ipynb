```python
def run_experiment(EXPERIMENT_SIZE, SEED):
    import jax
    import jax.numpy as jnp
    import optax

    from flax.training.train_state import TrainState
    from jax.flatten_util import ravel_pytree

    from src.config.core import Config
    from src.dataset import TextLoader
    from src.dataset.utils import CustomBPETokenizer
    import src.training.utils as train_utils
    from src.inference.evaluation import evaluate_bde

    print("\n" + "=" * 80)
    print(f"RUNNING {EXPERIMENT_SIZE} | SEED {SEED}")
    print("=" * 80)

    base_config_path = Path(CONFIGS[EXPERIMENT_SIZE])
    result_name = f"ag_news_mclmc_{EXPERIMENT_SIZE}_seed{SEED}"

    # --------------------------------------------------
    # 1. Create temporary seed config
    # --------------------------------------------------
    with open(base_config_path, "r") as f:
        cfg = yaml.safe_load(f)

    cfg["rng"] = SEED
    cfg["experiment_name"] = result_name

    tmp_config_path = MASTERARBEIT_DIR / "configs" / f"tmp_{result_name}.yaml"

    with open(tmp_config_path, "w") as f:
        yaml.safe_dump(cfg, f, sort_keys=False)

    print("Config:", tmp_config_path)
    print("Result:", result_name)

    # --------------------------------------------------
    # 2. Run MILE training
    # --------------------------------------------------
    cmd = [
        "python",
        "train.py",
        "-c",
        str(tmp_config_path),
        "-d",
        "1",
        "--device_limit",
        "1",
    ]

    print("\nRUNNING COMMAND:")
    print(" ".join(cmd))

    result = subprocess.run(
        cmd,
        cwd=str(MILE_DIR),
        text=True,
        capture_output=True,
    )

    print("\nSTDOUT:")
    print(result.stdout)

    print("\nSTDERR:")
    print(result.stderr)

    if result.returncode != 0:
        raise RuntimeError(f"train.py failed with return code {result.returncode}")

    # --------------------------------------------------
    # 3. Find newest timestamp folder and copy to Drive
    # --------------------------------------------------
    runs = sorted(
        [
            p for p in LOCAL_RESULTS.iterdir()
            if p.is_dir() and p.name.startswith(result_name)
        ],
        key=lambda p: p.stat().st_mtime,
    )

    if len(runs) == 0:
        raise FileNotFoundError(f"No local run found for {result_name}")

    latest_run = runs[-1]
    exp_dir = DRIVE_RESULTS / result_name

    if exp_dir.exists():
        shutil.rmtree(exp_dir)

    shutil.copytree(latest_run, exp_dir)

    if not (exp_dir / "samples").exists():
        raise FileNotFoundError("samples/ missing. MILE sampling probably failed.")

    print("Copied to:", exp_dir)

    # --------------------------------------------------
    # 4. Load model, data, samples
    # --------------------------------------------------
    config = Config.from_file(exp_dir / "config.yaml")
    tree_dir = exp_dir / "tree"

    module = config.get_flax_model()

    rng_eval = jax.random.key(SEED)

    tokenizer = CustomBPETokenizer(
        config.model.context_len,
        max_vocab_size=config.model.vocab_size,
    )

    loader = TextLoader(
        config=config.data,
        rng=rng_eval,
        tokenizer=tokenizer,
        context_len=config.model.context_len,
        omit_freq=100,
        target_len=config.data.target_len,
    )

    features = loader.test_x
    labels = loader.test_y

    print("Train:", loader.train_x.shape)
    print("Valid:", loader.valid_x.shape)
    print("Test :", features.shape)

    samples = train_utils.load_samples_from_dir(
        exp_dir / "samples",
        tree_path=tree_dir,
    )

    leaf = jax.tree.leaves(samples)[0]
    n_chains = leaf.shape[0]
    samples_per_chain = leaf.shape[1]
    posterior_samples = n_chains * samples_per_chain

    print("Chains:", n_chains)
    print("Samples per chain:", samples_per_chain)
    print("Posterior samples:", posterior_samples)

    # --------------------------------------------------
    # 5. Unified metric helpers
    # --------------------------------------------------
    def ece_from_probs(mean_probs, y, n_bins=15):
        confidences = jnp.max(mean_probs, axis=-1)
        predictions = jnp.argmax(mean_probs, axis=-1)
        accuracies = predictions == y

        ece = 0.0
        bin_edges = jnp.linspace(0.0, 1.0, n_bins + 1)

        for i in range(n_bins):
            lower = bin_edges[i]
            upper = bin_edges[i + 1]

            if i == 0:
                in_bin = (confidences >= lower) & (confidences <= upper)
            else:
                in_bin = (confidences > lower) & (confidences <= upper)

            prop_in_bin = jnp.mean(in_bin)

            bin_acc = jnp.where(
                jnp.sum(in_bin) > 0,
                jnp.mean(jnp.where(in_bin, accuracies, 0.0)) / prop_in_bin,
                0.0,
            )

            bin_conf = jnp.where(
                jnp.sum(in_bin) > 0,
                jnp.mean(jnp.where(in_bin, confidences, 0.0)) / prop_in_bin,
                0.0,
            )

            ece += prop_in_bin * jnp.abs(bin_acc - bin_conf)

        return ece

    def brier_from_probs(mean_probs, y):
        num_classes = mean_probs.shape[-1]
        y_onehot = jax.nn.one_hot(y, num_classes)
        return jnp.mean(jnp.sum((mean_probs - y_onehot) ** 2, axis=-1))

    def metrics_from_sample_probs(sample_probs, y):
        """
        sample_probs shape: (S, N, C)
        S = posterior samples
        N = observations
        C = classes
        """
        mean_probs = jnp.mean(sample_probs, axis=0)

        pred = jnp.argmax(mean_probs, axis=-1)
        acc = jnp.mean(pred == y)

        true_probs = mean_probs[jnp.arange(y.shape[0]), y]

        # Unified posterior predictive log likelihood
        lppd = jnp.mean(jnp.log(true_probs + 1e-12))
        nll = -lppd

        predictive_entropy = -jnp.mean(
            jnp.sum(mean_probs * jnp.log(mean_probs + 1e-12), axis=-1)
        )

        sample_entropies = -jnp.sum(
            sample_probs * jnp.log(sample_probs + 1e-12),
            axis=-1,
        )

        expected_entropy = jnp.mean(sample_entropies)

        mutual_information = predictive_entropy - expected_entropy

        ece = ece_from_probs(mean_probs, y)
        brier = brier_from_probs(mean_probs, y)

        return {
            "lppd": lppd,
            #"nll": nll,
            "acc": acc,
            "ece": ece,
            "brier": brier,
            "predictive_entropy": predictive_entropy,
            "mutual_information": mutual_information,
        }

    def metrics_from_logits(logits, y):
        probs = jax.nn.softmax(logits, axis=-1)

        if probs.ndim == 4:
            chains, samples_, N, C = probs.shape
            probs = probs.reshape(chains * samples_, N, C)

        return metrics_from_sample_probs(probs, y)

    # --------------------------------------------------
    # 6. MILE Evaluation
    # --------------------------------------------------
    mile_metrics_raw = {}

    mile_logits, _ = evaluate_bde(
        params=samples,
        module=module,
        features=features,
        labels=labels,
        task=config.data.task,
        batch_size=BATCH_SIZE_EVAL,
        verbose=False,
        metrics_dict=mile_metrics_raw,
        nominal_coverages=None,
    )

    mile_metrics = metrics_from_logits(mile_logits, labels)

    with open(exp_dir / f"mile_metrics_{EXPERIMENT_SIZE}_seed{SEED}.pkl", "wb") as f:
        pickle.dump(mile_metrics, f)

    # --------------------------------------------------
    # 7. MAP Training for MFVI/Laplace
    # --------------------------------------------------
    X_train = loader.train_x
    y_train = loader.train_y
    X_test = features
    y_test = labels

    rng = jax.random.key(SEED)
    rng, init_key = jax.random.split(rng)
    rng, train_key = jax.random.split(rng)
    rng, mfvi_key = jax.random.split(rng)
    rng, mfvi_sample_key = jax.random.split(rng)
    rng, laplace_sample_key = jax.random.split(rng)

    MAP_EPOCHS = config.training.warmstart.max_epochs
    MAP_BATCH_SIZE = config.training.warmstart.batch_size
    MAP_LR = config.training.warmstart.optimizer_config.parameters["learning_rate"]

    params_init = module.init(init_key, X_train[:1])

    def ce_loss(params, x, y):
        logits = module.apply(params, x, train=True)
        log_probs = jax.nn.log_softmax(logits, axis=-1)
        return -jnp.mean(log_probs[jnp.arange(y.shape[0]), y])

    @jax.jit
    def train_step(state, x, y):
        loss, grads = jax.value_and_grad(ce_loss)(state.params, x, y)
        return state.apply_gradients(grads=grads), loss

    state = TrainState.create(
        apply_fn=module.apply,
        params=params_init,
        tx=optax.adam(MAP_LR),
    )

    for epoch in range(MAP_EPOCHS):
        perm = jax.random.permutation(train_key, X_train.shape[0])
        train_key, _ = jax.random.split(train_key)

        Xs = X_train[perm]
        ys = y_train[perm]

        for start in range(0, X_train.shape[0], MAP_BATCH_SIZE):
            state, loss = train_step(
                state,
                Xs[start:start + MAP_BATCH_SIZE],
                ys[start:start + MAP_BATCH_SIZE],
            )

    map_params = state.params

    # --------------------------------------------------
    # 8. Last-layer setup
    # --------------------------------------------------
    def split_last_layer(params):
        p = params["params"]
        last_key = list(p.keys())[-1]
        base = {k: v for k, v in p.items() if k != last_key}
        last = {last_key: p[last_key]}
        return {"params": base}, {"params": last}, last_key

    base_params, last_params, last_key = split_last_layer(map_params)
    flat_last, unravel_last = ravel_pytree(last_params)

    D = flat_last.shape[0]

    def combine_params(base_params, last_params):
        return {
            "params": {
                **base_params["params"],
                **last_params["params"],
            }
        }

    def forward_from_vec(theta_vec, x):
        last_p = unravel_last(theta_vec)
        full_p = combine_params(base_params, last_p)
        return module.apply(full_p, x, train=False)

    @jax.jit
    def probs_for_theta(theta_vec, x):
        logits = forward_from_vec(theta_vec, x)
        return jax.nn.softmax(logits, axis=-1)

    # --------------------------------------------------
    # 9. MFVI
    # --------------------------------------------------
    def log_prior(theta_vec, prior_scale=1.0):
        return -0.5 * jnp.sum((theta_vec / prior_scale) ** 2)

    def log_likelihood(theta_vec, x, y):
        logits = forward_from_vec(theta_vec, x)
        log_probs = jax.nn.log_softmax(logits, axis=-1)
        return jnp.sum(log_probs[jnp.arange(y.shape[0]), y])

    def elbo(var_params, x, y, rng, num_samples=8):
        mean = var_params["mean"]
        log_std = var_params["log_std"]
        std = jnp.exp(log_std)

        def one_sample(key):
            eps = jax.random.normal(key, mean.shape)
            theta = mean + std * eps

            ll = log_likelihood(theta, x, y)
            lp = log_prior(theta)

            dim = mean.shape[0]
            lq = (
                -0.5 * jnp.sum(((theta - mean) / std) ** 2)
                - jnp.sum(log_std)
                - 0.5 * dim * jnp.log(2.0 * jnp.pi)
            )

            return ll + lp - lq

        keys = jax.random.split(rng, num_samples)
        return jnp.mean(jax.vmap(one_sample)(keys))

    MFVI_EPOCHS = 500
    MFVI_BATCH_SIZE = MAP_BATCH_SIZE
    MFVI_LR = MAP_LR
    MFVI_ELBO_SAMPLES = 8

    optimizer = optax.adam(MFVI_LR)

    var_params = {
        "mean": flat_last,
        "log_std": jnp.full_like(flat_last, -5.0),
    }

    opt_state = optimizer.init(var_params)

    @jax.jit
    def mfvi_step(var_params, opt_state, x, y, rng):
        loss_fn = lambda vp: -elbo(
            vp,
            x,
            y,
            rng,
            num_samples=MFVI_ELBO_SAMPLES,
        )
        loss, grads = jax.value_and_grad(loss_fn)(var_params)
        updates, opt_state = optimizer.update(grads, opt_state)
        var_params = optax.apply_updates(var_params, updates)
        return var_params, opt_state, loss

    for epoch in range(MFVI_EPOCHS):
        perm = jax.random.permutation(mfvi_key, X_train.shape[0])
        mfvi_key, _ = jax.random.split(mfvi_key)

        Xs = X_train[perm]
        ys = y_train[perm]

        for start in range(0, X_train.shape[0], MFVI_BATCH_SIZE):
            mfvi_key, subkey = jax.random.split(mfvi_key)
            var_params, opt_state, loss = mfvi_step(
                var_params,
                opt_state,
                Xs[start:start + MFVI_BATCH_SIZE],
                ys[start:start + MFVI_BATCH_SIZE],
                subkey,
            )

    eps = jax.random.normal(
        mfvi_sample_key,
        (posterior_samples, D),
    )

    theta_samples_mfvi = (
        var_params["mean"][None, :]
        + jnp.exp(var_params["log_std"])[None, :] * eps
    )

    sample_probs_mfvi = jax.vmap(
        lambda th: probs_for_theta(th, X_test)
    )(theta_samples_mfvi)

    mfvi_metrics = metrics_from_sample_probs(sample_probs_mfvi, y_test)

    with open(exp_dir / f"mfvi_last_layer_metrics_{EXPERIMENT_SIZE}_seed{SEED}.pkl", "wb") as f:
        pickle.dump(mfvi_metrics, f)

    # --------------------------------------------------
    # 10. Laplace
    # --------------------------------------------------
    def grad_logits(theta_vec, x_single):
        def logits_fn(th):
            return forward_from_vec(th, x_single[None, :])[0]
        return jax.jacrev(logits_fn)(theta_vec)

    def softmax_hessian_logits(theta_vec, x_single):
        logits = forward_from_vec(theta_vec, x_single[None, :])[0]
        p = jax.nn.softmax(logits)
        return jnp.diag(p) - jnp.outer(p, p)

    @jax.jit
    def ggn_one(theta_vec, x_single):
        J = grad_logits(theta_vec, x_single)
        H = softmax_hessian_logits(theta_vec, x_single)
        return J.T @ H @ J

    def compute_laplace_precision(theta_map, X, prior_precision=1.0):
        precision = prior_precision * jnp.eye(theta_map.shape[0])

        for i in range(X.shape[0]):
            precision = precision + ggn_one(theta_map, X[i])

        return precision

    theta_map = flat_last

    laplace_precision = compute_laplace_precision(
        theta_map=theta_map,
        X=X_train,
        prior_precision=1.0,
    )

    def sample_laplace(theta_map, precision, rng, n_samples, jitter=1e-3):
        D = theta_map.shape[0]

        precision_jitter = precision + jitter * jnp.eye(D)
        L = jnp.linalg.cholesky(precision_jitter)

        eps = jax.random.normal(rng, (n_samples, D))

        z = jax.scipy.linalg.solve_triangular(
            L.T,
            eps.T,
            lower=False,
        ).T

        return theta_map[None, :] + z

    theta_samples_laplace = sample_laplace(
        theta_map=theta_map,
        precision=laplace_precision,
        rng=laplace_sample_key,
        n_samples=posterior_samples,
        jitter=1e-3,
    )

    sample_probs_laplace = jax.vmap(
        lambda th: probs_for_theta(th, X_test)
    )(theta_samples_laplace)

    laplace_metrics = metrics_from_sample_probs(sample_probs_laplace, y_test)

    with open(exp_dir / f"laplace_last_layer_metrics_{EXPERIMENT_SIZE}_seed{SEED}.pkl", "wb") as f:
        pickle.dump(laplace_metrics, f)

    # --------------------------------------------------
    # 11. Combined CSV
    # --------------------------------------------------
    rows = []

    for method_name, metrics in [
        ("MILE", mile_metrics),
        ("Last-Layer MFVI", mfvi_metrics),
        ("Last-Layer Laplace", laplace_metrics),
    ]:
        row = {
            "size": EXPERIMENT_SIZE,
            "seed": SEED,
            "method": method_name,
            "n_chains": int(n_chains),
            "samples_per_chain": int(samples_per_chain),
            "posterior_samples": int(posterior_samples),
        }

        for k, v in metrics.items():
            try:
                row[k] = float(v)
            except Exception:
                row[k] = v

        rows.append(row)

    run_df = pd.DataFrame(rows)

    csv_path = exp_dir / f"comparison_{EXPERIMENT_SIZE}_seed{SEED}.csv"
    pkl_path = exp_dir / f"comparison_{EXPERIMENT_SIZE}_seed{SEED}.pkl"

    run_df.to_csv(csv_path, index=False)

    with open(pkl_path, "wb") as f:
        pickle.dump(run_df, f)

    print("\nFinished:")
    print(run_df)

    return run_df
```
